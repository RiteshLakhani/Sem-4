```python
import matplotlib.pyplot as plt

# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

for i, size in enumerate(sizes):
    percentage = sizes[i] / sum(sizes) * 100
    plt.text(x=0.5, y=0.5, s=f'{labels[i]}: {percentage:.1f}%', horizontalalignment='center', verticalalignment='center')



# Title
plt.title('Pie Chart with Annotations')

# Show the plot
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import matplotlib.pyplot as plt
import random

num_points = 20

x_coords = [random.randint(0, 100) for _ in range(num_points)]
y_coords = [random.randint(0, 100) for _ in range(num_points)]

plt.figure(figsize=(8,6))
plt.scatter(x_coords, y_coords, color = 'red')
plt.title('Random X, Y coordinates')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt

# Sample data
savings = [20, 30, 40, 50]
num_employees = [105, 199, 29, 73]
colors = ['red', 'green', 'yellow', 'cyan']
w = [0.9,0.6,0.7,0.8]

# Create bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(savings, num_employees, color=colors,width = w)

# Add text labels on top of each bar
for bar, num_employee in zip(bars, num_employees):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, num_employee, ha='center', va='bottom')

# Add labels and title
plt.xlabel('Saving (%)')
plt.ylabel('Number of Employees')
plt.title('Monthly Salary Savings by Employees')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['101', '102', '103', '104']
values = [25, 40, 30, 35]


plt.figure(figsize=(8,6))
bars = plt.bar(categories, values, color='skyblue')
for bar,values in zip(bars, values):
    height = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,height,values,ha='center',va='bottom')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for two lines
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create line chart
plt.figure(figsize=(8, 6))

# Plotting the first line with customization
plt.plot(x, y1, color='blue', linewidth=2, linestyle='--', marker='o', label='Sin(x)')

# Plotting the second line with customization
plt.plot(x, y2, color='red', linewidth=1.5, linestyle='-.', marker='s', label='Cos(x)')

# Adding annotations
plt.text(2, 0.5, 'Example Annotation', fontsize=12, color='green')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart with Two Lines')

# Adding legend
plt.legend(loc=5)

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['101', '102', '103', '104']
values = [25, 40, 30, 35]

plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values, color='skyblue')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, '%d' % int(height), ha='center', va='bottom')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar chart Example with value Labels')
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)  

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, cmap='viridis', alpha=0.7)  
plt.colorbar()  
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Different Colors')
plt.grid(True)
plt.show()
```
